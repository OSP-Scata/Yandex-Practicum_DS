```python
import pandas as pd
```

```python
pps = pd.read_excel('–ü–µ–Ω–æ–ø–ª–∞—Å—Ç –º–æ–¥–µ—Ä–∞—Ü–∏—è.xlsx', sheet_name='–°–µ–Ω—Ç—è–±—Ä—å_2023')
pps.head()
```

```python
pps = pps.drop(['Unnamed: 8', '–∫–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π Textarium', '–°—Ç—Ä–æ–∫', '—Ç–µ–º–∞—Ç–∏–∫–∞'], axis=1)
pps.head()
```

```python
# decade = pps.loc[(pps['–¥–∞—Ç–∞'] >= '2023-09-01') & (pps['–¥–∞—Ç–∞'] <= '2023-09-10')]
# decade = pps.loc[(pps['–¥–∞—Ç–∞'] >= '2023-09-11') & (pps['–¥–∞—Ç–∞'] <= '2023-09-20')]
decade = pps.loc[(pps['–¥–∞—Ç–∞'] >= '2023-09-21') & (pps['–¥–∞—Ç–∞'] <= pps['–¥–∞—Ç–∞'].max())]
decade.head()
```

```python
month = pps.dropna(subset=['—Ç–µ–∫—Å—Ç –æ—Ç–≤–µ—Ç–∞'])
neg_month = month.loc[pps['—Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å'] == '–Ω–µ–≥–∞—Ç–∏–≤']
neu_month = month.loc[pps['—Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å'] == '–Ω–µ–π—Ç—Ä–∞–ª']
pos_month = month.loc[pps['—Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å'] == '–ø–æ–∑–∏—Ç–∏–≤']

m_neg_count = neg_month['—Å–∫—Ä–∏–Ω—à–æ—Ç'].str.split('\n', expand=True).notna().sum().sum()
m_pos_count = pos_month['—Å–∫—Ä–∏–Ω—à–æ—Ç'].str.split('\n', expand=True).notna().sum().sum()
m_neu_count = neu_month['—Å–∫—Ä–∏–Ω—à–æ—Ç'].str.split('\n', expand=True).notna().sum().sum()

print('–ù–µ–≥–∞—Ç–∏–≤ –º–µ—Å—è—Ü', m_neg_count, '\n–ü–æ–∑–∏—Ç–∏–≤ –º–µ—Å—è—Ü', m_pos_count, '\n–ù–µ–π—Ç—Ä–∞–ª –º–µ—Å—è—Ü', m_neu_count, '\n–í—Å–µ–≥–æ –º–µ—Å—è—Ü', 
      m_neg_count + m_pos_count + m_neu_count)
```

```python
negative = decade.loc[pps['—Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å'] == '–Ω–µ–≥–∞—Ç–∏–≤'].dropna(subset=['—Ç–µ–∫—Å—Ç –æ—Ç–≤–µ—Ç–∞']).dropna(subset=['—Å–∫—Ä–∏–Ω—à–æ—Ç'])
count_neg = negative['—Å–∫—Ä–∏–Ω—à–æ—Ç'].str.split('\n', expand=True).notna().sum().sum()
negative.head()
```

```python
positive = decade.loc[pps['—Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å'] == '–ø–æ–∑–∏—Ç–∏–≤'].dropna(subset=['—Ç–µ–∫—Å—Ç –æ—Ç–≤–µ—Ç–∞']).dropna(subset=['—Å–∫—Ä–∏–Ω—à–æ—Ç'])
count_pos = positive['—Å–∫—Ä–∏–Ω—à–æ—Ç'].str.split('\n', expand=True).notna().sum().sum()
```

```python
neutral = decade.loc[pps['—Ç–æ–Ω–∞–ª—å–Ω–æ—Å—Ç—å'] == '–Ω–µ–π—Ç—Ä–∞–ª'].dropna(subset=['—Ç–µ–∫—Å—Ç –æ—Ç–≤–µ—Ç–∞']).dropna(subset=['—Å–∫—Ä–∏–Ω—à–æ—Ç'])
count_neu = neutral['—Å–∫—Ä–∏–Ω—à–æ—Ç'].str.split('\n', expand=True).notna().sum().sum()
```

```python
print('–ù–µ–≥–∞—Ç–∏–≤', count_neg, '\n–ü–æ–∑–∏—Ç–∏–≤', count_pos, '\n–ù–µ–π—Ç—Ä–∞–ª', count_neu, '\n–í—Å–µ–≥–æ', count_neg+count_pos+count_neu)
```
