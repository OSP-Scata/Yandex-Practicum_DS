```python
import pandas as pd
```

```python
pps = pd.read_excel('Пенопласт модерация.xlsx', sheet_name='Сентябрь_2023')
pps.head()
```

```python
pps = pps.drop(['Unnamed: 8', 'комментарий Textarium', 'Строк', 'тематика'], axis=1)
pps.head()
```

```python
# decade = pps.loc[(pps['дата'] >= '2023-09-01') & (pps['дата'] <= '2023-09-10')]
# decade = pps.loc[(pps['дата'] >= '2023-09-11') & (pps['дата'] <= '2023-09-20')]
decade = pps.loc[(pps['дата'] >= '2023-09-21') & (pps['дата'] <= pps['дата'].max())]
decade.head()
```

```python
month = pps.dropna(subset=['текст ответа'])
neg_month = month.loc[pps['тональность'] == 'негатив']
neu_month = month.loc[pps['тональность'] == 'нейтрал']
pos_month = month.loc[pps['тональность'] == 'позитив']

m_neg_count = neg_month['скриншот'].str.split('\n', expand=True).notna().sum().sum()
m_pos_count = pos_month['скриншот'].str.split('\n', expand=True).notna().sum().sum()
m_neu_count = neu_month['скриншот'].str.split('\n', expand=True).notna().sum().sum()

print('Негатив месяц', m_neg_count, '\nПозитив месяц', m_pos_count, '\nНейтрал месяц', m_neu_count, '\nВсего месяц', 
      m_neg_count + m_pos_count + m_neu_count)
```

```python
negative = decade.loc[pps['тональность'] == 'негатив'].dropna(subset=['текст ответа']).dropna(subset=['скриншот'])
count_neg = negative['скриншот'].str.split('\n', expand=True).notna().sum().sum()
negative.head()
```

```python
positive = decade.loc[pps['тональность'] == 'позитив'].dropna(subset=['текст ответа']).dropna(subset=['скриншот'])
count_pos = positive['скриншот'].str.split('\n', expand=True).notna().sum().sum()
```

```python
neutral = decade.loc[pps['тональность'] == 'нейтрал'].dropna(subset=['текст ответа']).dropna(subset=['скриншот'])
count_neu = neutral['скриншот'].str.split('\n', expand=True).notna().sum().sum()
```

```python
print('Негатив', count_neg, '\nПозитив', count_pos, '\nНейтрал', count_neu, '\nВсего', count_neg+count_pos+count_neu)
```
