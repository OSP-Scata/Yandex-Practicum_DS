### Итоги открытия данных
Столбец даты во всех наборах содержит строку (object), а не тип datetime, нужно привести к правильному формату. Также в данных есть пропуски, но столбцов слишком много, чтобы их количество можно было оценить на глаз. Посчитаем сумму.

```python
train['date'] = pd.to_datetime(train['date'])
test['date'] = pd.to_datetime(test['date'])
full['date'] = pd.to_datetime(full['date'])

df_array = [train, test, full]

for i in range(len(df_array)):
    print('Набор данных №', i)
    # print('Пропуски по столбцам:', df_array[i].isna().sum()) # на всякий случай
    print('Всего пропусков:', df_array[i].isna().sum().sum())
    print()
```

Теперь проверим, что эффективность обогащения рассчитана правильно, и посчитаем МАЕ.

```python
train.dropna(subset=['rougher.output.concentrate_au', 'rougher.input.feed_au',
                     'rougher.output.tail_au','rougher.output.recovery'], inplace=True)
# 
c = train['rougher.output.concentrate_au']
f = train['rougher.input.feed_au']
t = train['rougher.output.tail_au']
recovery_true = train['rougher.output.recovery']
recovery_calculated = (c * (f - t)) / (f * (c - t)) * 100
print('MAE =', mean_absolute_error(recovery_true, recovery_calculated))
```

МАЕ имеет весьма малую величину, а значит, рассчитанное значение совпадает с данными, записанными в столбце `rougher.output.recovery`. Теперь удаляем все пропуски в обучающей и тестовой выборке.

```python
train_cleaned = train.dropna()
test_cleaned = test.dropna()
print(train_cleaned.isna().sum().sum(), test_cleaned.isna().sum().sum())
```

## Модель

### Метрика sMAPE
Сначала пишем функцию для вычисления sMAPE.

```python
# сначала я попробовала реализовать так, но не получилось
'''def sMAPE(rougher_recovery_true, rougher_recovery_pred, final_recovery_true, final_recovery_pred):
    n = rougher_recovery_true.shape[0]
    rougher_sMAPE = ((2 * abs(rougher_recovery_true - rougher_recovery_pred)/
                     (abs(rougher_recovery_true) + abs(rougher_recovery_pred))).sum() * 100 / n)
    final_sMAPE = ((2*abs(final_recovery_true - final_recovery_pred)/
                     (abs(final_recovery_true) + abs(final_recovery_pred))).sum() * 100 / n)
    return rougher_sMAPE * 0.25 + final_sMAPE * 0.75
    '''
def sMAPE(true, pred):
    error = np.mean(abs(true - pred)/((abs(true) + abs(pred)) / 2)) * 100
    return error

def final_sMAPE(error_rougher, error_final):
    return (0.25 * error_rougher + 0.75 * error_final)

custom_score = make_scorer(sMAPE, greater_is_better=False)
```

### Модели
Поскольку у нас задача регрессии, выбираем модели, заточенные под эту работу.

```python
linear = make_pipeline(StandardScaler(), LinearRegression())
tree = make_pipeline(StandardScaler(), DecisionTreeRegressor())
forest = make_pipeline(StandardScaler(), RandomForestRegressor())
```

```python
# параметры для SearchGrid - дерево и лес
params_tree = {'decisiontreeregressor__max_depth':[1,10]}
params_forest = {'randomforestregressor__n_estimators':[5,100],
             'randomforestregressor__max_depth':[1,10]}
```

```python
# модели, которые будут задаваться и обучаться через параметры SearchGrid
grid_rougher_tree = GridSearchCV(tree, param_grid=params_tree, scoring=custom_score, error_score='raise')
grid_final_tree = GridSearchCV(tree, param_grid=params_tree, scoring=custom_score, error_score='raise')
grid_rougher_forest = GridSearchCV(forest, param_grid=params_forest, scoring=custom_score, error_score='raise')
grid_final_forest = GridSearchCV(forest, param_grid=params_forest, scoring=custom_score, error_score='raise')
```

```python
# обучение
grid_rougher_tree.fit(train_features_rougher, train_target_rougher)
grid_final_tree.fit(train_features_final, train_target_final)

grid_rougher_forest.fit(train_features_rougher, train_target_rougher)
grid_final_forest.fit(train_features_final, train_target_final)
```

Определим лучшую модель.

```python
print('Лучшие модели:')
print(grid_rougher_tree.best_estimator_)
print(grid_final_tree.best_estimator_)
print(grid_rougher_forest.best_estimator_)
print(grid_final_forest.best_estimator_)
```

```python
bestscore_rougher_tree = grid_rougher_tree.best_score_
bestscore_final_tree = grid_final_tree.best_score_
bestscore_rougher_forest = grid_rougher_forest.best_score_
bestscore_final_forest = grid_final_forest.best_score_

# в дерево и лес уже встроена кросс-валидация
print('Этап rougher:')
print('DecisionTree:', bestscore_rougher_tree, 'RandomForest:', bestscore_rougher_forest)

print('Этап final:')
print('DecisionTree:', bestscore_rougher_tree, 'RandomForest:', bestscore_rougher_forest)
```

Метрики линейной регрессии проверяем отдельно, т.к. у неё нет встроенной кросс-валидации, позволяющей использовать `best_score_`.

```python
print('Кросс-валидация LinearRegression, средняя R2:', cross_val_score(linear, train_features_final, train_target_final,
                                                                       cv=5).mean())
print('Кросс-валидация LinearRegression, средняя МАЕ:', cross_val_score(linear, train_features_final, train_target_final,
                                                                        cv=5, scoring='neg_mean_absolute_error'))
```

sMAPE для дерева регрессии:

```python
final_sMAPE(bestscore_rougher_tree, bestscore_final_tree)
```

sMAPE для леса:

```python
final_sMAPE(bestscore_rougher_forest, bestscore_final_forest)
```

В целом лучшие результаты даёт случайный лес, поэтому проверять на тестовой выборке будем его.

```python
test_rougher = grid_rougher_forest.predict(test_cleaned)
test_final = grid_final_forest.predict(test_cleaned)
```

```python
final_sMAPE(sMAPE(target_test_rougher, test_rougher), sMAPE(target_test_final, test_final))
```

Инициализируем константную модель для проверки адекватности.

```python
dummy_rougher = DummyRegressor(strategy='median')
dummy_final = DummyRegressor(strategy='median')

dummy_rougher.fit(train_features_rougher, train_target_rougher)
dummy_final.fit(train_features_final, train_target_final)
```

```python
dummy_pred_rougher = dummy_rougher.predict(test_cleaned)
dummy_pred_final = dummy_final.predict(test_cleaned)

final_sMAPE(sMAPE(target_test_rougher, dummy_pred_rougher), sMAPE(target_test_final, dummy_pred_final))
```
