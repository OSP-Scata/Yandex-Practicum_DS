–†–∞–¥–∏ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞ –¥–∞—Ç–∞—Å–µ—Ç `data_clean` –æ—Å—Ç–∞–≤–∏–º –Ω–µ–∏–∑–º–µ–Ω–Ω—ã–º, –Ω–æ —Å–æ–∑–¥–∞–¥–∏–º –≤—Ç–æ—Ä–æ–π - `data_cleaner`, –∏–∑ –∫–æ—Ç–æ—Ä–æ–≥–æ —É–±–µ—Ä—ë–º —Ç–∞–∫–∂–µ —Å—Ç–æ–ª–±—Ü—ã —Å –¥–æ–ø.—É—Å–ª—É–≥–∞–º–∏ (–æ—Å—Ç–∞–≤–∏–º —Ç–æ–ª—å–∫–æ –æ–±—â–∏–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∏ Internet/Phone). –ü—Ä–æ—Ç–µ—Å—Ç–∏—Ä—É–µ–º –æ–±–∞.

## –°–æ–∑–¥–∞–Ω–∏–µ –≤—ã–±–æ—Ä–æ–∫, –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π
### –†–∞–∑–±–∏–µ–Ω–∏–µ –Ω–∞ –≤—ã–±–æ—Ä–∫–∏, –û–ù–ï
–°–æ–∑–¥–∞—ë–º —Å–ø–∏—Å–∫–∏ —á–∏—Å–ª–µ–Ω–Ω—ã—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤.

```python
numeric = ['MonthlyCharges', 'Duration']
categorical = ['Type', 'PaperlessBilling', 'PaymentMethod', 'SeniorCitizen', 'Partner', 'Dependents', 'Internet', 
               'Phone', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 
               'StreamingTV', 'StreamingMovies', 'MultipleLines']
categorical1 = ['Type', 'PaperlessBilling', 'PaymentMethod', 'SeniorCitizen', 'Partner', 'Dependents', 'Internet', 
               'Phone']
```

```python
# –Ω–∞–∑–Ω–∞—á–µ–Ω–∏–µ —Ç–∏–ø–∞ category –¥–ª—è LGBM/CatBoost
for col in categorical:
    data_clean[col] = data_clean[col].astype('category')
    
for col in categorical1:
    data_cleaner[col] = data_cleaner[col].astype('category')
```

–í—ã–¥–µ–ª—è–µ–º –æ–±—É—á–∞—é—â–∏–µ –∏ —Ü–µ–ª–µ–≤–æ–π –ø—Ä–∏–∑–Ω–∞–∫–∏.

```python
target = data_clean['ClientLeft']

features = data_clean.drop('ClientLeft', axis=1)
features1 = data_cleaner.drop('ClientLeft', axis=1)
```

–†–∞–∑–¥–µ–ª—è–µ–º –æ–±–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏, –ø–æ—Å–ª–µ —Ä–∞–∑–±–∏–µ–Ω–∏—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –ø–µ—Ä–µ–∫–æ–¥–∏—Ä—É–µ–º –ø–æ OneHotEncoding.

```python
# –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –æ—Å–Ω–æ–≤–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=TEST_SIZE, random_state=RANDOM_STATE)

# –≤—ã–±–æ—Ä–∫–∏ –∏–∑ –±–æ–ª–µ–µ —á–∏—Å—Ç–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞ - –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å, —á—Ç–æ –±—É–¥–µ—Ç –ª—É—á—à–µ
X1_train, X1_test, y1_train, y1_test = train_test_split(features1, target, test_size=TEST_SIZE, random_state=RANDOM_STATE)

X_train_ohe = X_train.copy() # –æ—Ç–¥–µ–ª—å–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏ –¥–ª—è –û–ù–ï-–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏—è
X_test_ohe = X_test.copy()

X1_train_ohe = X1_train.copy()
X1_test_ohe = X1_test.copy()
```

–í—ã–¥–µ–ª—è–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∏ —á–∏—Å–ª–µ–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏, –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –∫–æ–¥–∏—Ä—É–µ–º –≤ –û–ù–ï, –∞ —á–∏—Å–ª–µ–Ω–Ω—ã–µ –º–∞—Å—à—Ç–∞–±–∏—Ä—É–µ–º. –ü–æ—Å–∫–æ–ª—å–∫—É —É –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–æ–≤–æ–ª—å–Ω–æ –º–∞–ª–æ "–∫–∞—Ç–µ–≥–æ—Ä–∏–π", –¥–æ–ª–∂–Ω–æ —Ö–≤–∞—Ç–∏—Ç—å –æ–±—ã—á–Ω–æ–≥–æ OneHotEncoder, –±–µ–∑ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ—Å—Ç–∏ OrdinalEncoder.

```python
encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)

#–æ–±—É—á–∞–µ–º —ç–Ω–∫–æ–¥–µ—Ä –Ω–∞ –∑–∞–¥–∞–Ω–Ω—ã—Ö –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏
encoder_ohe.fit(X_train_ohe[categorical])

#–¥–æ–±–∞–≤–ª—è–µ–º –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –≤ X_train_ohe
#encoder_ohe.get_feature_names_out() –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–æ–ª—É—á–∏—Ç—å –Ω–∞–∑–≤–∞–Ω–∏—è –∫–æ–ª–æ–Ω–æ–∫
X_train_ohe[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(X_train_ohe[categorical])

#—É–¥–∞–ª—è–µ–º –Ω–µ–∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ (–∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–µ –∫–æ–ª–æ–Ω–∫–∏)
X_train_ohe = X_train_ohe.drop(categorical, axis=1)

#—Å–æ–∑–¥–∞—ë–º —Å–∫–µ–π–ª–µ—Ä
scaler = StandardScaler()

#–æ–±—É—á–∞–µ–º –µ–≥–æ –Ω–∞ —á–∏—Å–ª–µ–Ω–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏, —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º –µ—ë –∂–µ
X_train_ohe[numeric] = scaler.fit_transform(X_train_ohe[numeric])

#—Å–º–æ—Ç—Ä–∏–º –Ω–∞ —Ä–µ–∑—É–ª—å—Ç–∞—Ç
X_train_ohe.head()
```

```python
# —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–∏—Ä—É–µ–º —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫—É —É–∂–µ –±–µ–∑ –æ–±—É—á–µ–Ω–∏—è
X_test_ohe[
    encoder_ohe.get_feature_names_out()
] = encoder_ohe.transform(X_test_ohe[categorical])
X_test_ohe = X_test_ohe.drop(categorical, axis=1)
X_test_ohe[numeric] = scaler.transform(X_test_ohe[numeric])

X_test_ohe.head()
```

### –°–æ–∑–¥–∞–Ω–∏–µ –∏ –æ–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π
–°–æ–∑–¥–∞–¥–∏–º –º–æ–¥–µ–ª–∏, —Å –ø–æ–º–æ—â—å—é –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –ø–æ–¥–±–µ—Ä—ë–º –ª—É—á—à—É—é. –ë—É–¥–µ–º –ø—Ä–æ–±–æ–≤–∞—Ç—å –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é, —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å, CatBoost –∏ LightGBM.

```python
model_logistic = LogisticRegression(random_state=RANDOM_STATE, class_weight='balanced')
model_forest = RandomForestClassifier(random_state=RANDOM_STATE, class_weight='balanced')

# CatBoost –∏ LightGBM –±—É–¥–µ–º —É—á–∏—Ç—å –Ω–∞ –≤–∏–¥–µ–æ–∫–∞—Ä—Ç–µ
model_lgbm = LGBMClassifier(random_state=RANDOM_STATE, class_weight='balanced', objective='binary',
                            metric='binary_logloss', device='gpu', verbose=-1, gpu_use_dp=False, max_bin=63)
model_cat = CatBoostClassifier(cat_features=categorical, random_state=RANDOM_STATE, auto_class_weights='Balanced',
                              task_type="GPU", devices='0:1', gpu_ram_part=0.9, verbose=0)
```

–ü–æ–¥–±–∏—Ä–∞–µ–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã —Å –ø–æ–º–æ—â—å—é GridSearchCV.

–õ–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è:

```python
param_grid_logistic = {
    'solver': ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
}

gs_logistic = GridSearchCV(
    model_logistic, 
    param_grid=param_grid_logistic, 
    scoring='roc_auc', 
    n_jobs=-1
)

gs_logistic.fit(X_train_ohe, y_train)

print(f'best_score: {gs_logistic.best_score_}')

print(f'best_params: {gs_logistic.best_params_}')
```

–°–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å:

```python
param_grid_rf = {
    'n_estimators': range(50, 301, 50),
    'max_depth': range(2, 20),
    'min_samples_split': (2, 3, 4, 5, 6),
    'min_samples_leaf': (1, 2, 3, 4, 5)
}

gs_rf = RandomizedSearchCV(
    model_forest,
    param_distributions=param_grid_rf,
    scoring='roc_auc',
    n_jobs=-1,
    random_state=RANDOM_STATE
)

gs_rf.fit(X_train_ohe, y_train)

gs_rf_best_score = gs_rf.best_score_
gs_rf_best_params = gs_rf.best_params_
print(f'best_score: {gs_rf_best_score}')
print(f'best_params: {gs_rf_best_params}')
```

CatBoost:

```python
cat_parameters = {'depth': [3, 5, 7, 9, 11],
            'learning_rate': [0.5, 0.1, 0.05, 0.01],
            'iterations': [50, 100, 150, 200, 250]
}

grid_cat = GridSearchCV(model_cat, param_grid=cat_parameters, cv=2, n_jobs=None, scoring='roc_auc')
grid_cat.fit(X_train, y_train)
cat_best_score = grid_cat.best_score_
cat_best_params = grid_cat.best_params_
print("The best score across ALL searched params:", cat_best_score)
print("The best parameters across ALL searched params:", cat_best_params)
```

```python
lgbm_parameters = {
    'max_depth': [3, 5, 7, 9, 11], 
    'learning_rate': [0.5, 0.1, 0.05, 0.01],
    'num_leaves': [15, 31, 63],
    'n_estimators': [50, 100, 150, 200, 250]
}

grid_lgbm = GridSearchCV(
    estimator=model_lgbm,
    param_grid=lgbm_parameters,
    cv=5,
    n_jobs=None
)
grid_lgbm.fit(X_train, y_train)
lgbm_best_score = grid_lgbm.best_score_
lgbm_best_params = grid_lgbm.best_params_
print("The best score across ALL searched params:", lgbm_best_score)
print("The best parameters across ALL searched params:", lgbm_best_params)
```
